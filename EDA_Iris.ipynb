## Introduction

The Iris dataset is a classic dataset in the field of machine learning and data analysis. It contains measurements of various attributes of three species of iris flowers: Setosa, Versicolor, and Virginica. In this notebook, we will conduct an exploratory data analysis (EDA) to understand the relationships between different attributes and species of iris flowers.


## Dataset Overview

Let's start by loading the dataset and exploring its structure.

```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns

# Loading the Iris dataset
iris_df = sns.load_dataset('iris')

print("First few rows of the dataset:")
# Displaying the first few rows of the dataset
iris_df.head()
```

```python
# Display dataset information
print("\nDataset information:")
iris_df.info()
```

The Iris dataset consists of 150 samples, with each sample having four features (sepal length, sepal width, petal length, and petal width) and one target variable (species). The target variable 'species' has three classes: Setosa, Versicolor, and Virginica.

## Exploratory Data Analysis:

### 1. Dataset Exploration:
We started the analysis by exploring the dataset. The dataset comprises four numerical features and one categorical feature. The numerical features include sepal length, sepal width, petal length, and petal width, while the categorical feature is the species of iris flowers.

### 2. Visualizing Categorical Data

We'll visualize the distribution of iris species using a bar chart.

```python
import matplotlib.pyplot as plt

# Plotting the distribution of iris species
plt.figure(figsize=(8, 6))
sns.countplot(x='species', data=iris_df)
plt.title('Distribution of Iris Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()
```

We visualized the distribution of iris species using a bar chart. The distribution is fairly balanced, with each species (Setosa, Versicolor, and Virginica) having an equal representation of 50 samples in the dataset.



### 3. Visualizing Numerical Data

Next, let's explore the relationships between numerical features using scatter plots.

```python
# Plotting petal length against sepal length
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_length', y='petal_length', data=iris_df)
plt.title('Petal Length vs Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.show()
```

Next, we explored the relationships between numerical features using scatter plots. We plotted the petal length against the sepal length to observe any patterns. The scatter plot revealed clear clusters of data points, indicating potential correlations between these features.


### 4. Visualizing Relationships between Variables

To gain further insights, we'll plot a pairplot to visualize relationships between variables.

```python
# Plotting a pairplot
plt.figure(figsize=(10, 8))
sns.pairplot(iris_df, hue='species')
plt.title('Pairplot of Iris Dataset')
plt.show()
```

To gain further insights into the relationships between variables, we plotted a pairplot. The pairplot displays scatter plots for all pairs of numerical features, along with histograms on the diagonal. Additionally, different species are represented by different colors, allowing us to observe species-specific patterns.

### 5. Statistics

Finally, let's compute descriptive statistics and the correlation matrix for the dataset.

```python
# Descriptive statistics
iris_stats = iris_df.describe()
iris_stats
```

```python
# Correlation matrix
corr_matrix = iris_df.drop('species', axis=1).corr()
corr_matrix
```

We computed descriptive statistics and correlation matrix for the dataset. Descriptive statistics provided insights into the central tendency, dispersion, and shape of the numerical features. The correlation matrix revealed the degree of linear relationship between pairs of features. Strong positive correlations were observed between petal length and petal width, as well as between sepal length and petal length.

The exploratory data analysis of the Iris dataset provided valuable insights into the relationships between different attributes and species of iris flowers. The visualizations helped us understand patterns and distributions within the dataset, while descriptive statistics and correlation analysis quantified the relationships between numerical features.
